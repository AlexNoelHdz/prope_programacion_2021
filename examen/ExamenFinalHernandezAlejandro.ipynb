**2.2.** Realizar el merge entre las tablas "user_usage" y "user_device", de manera que se preserven todos los datos de "user_usage". 

Cuidar de no incluir en el merge variables innecesarias que no tengan relación con el problema. (15 puntos)

```python
# Merge entre las tablas user_usage y user_device
#Left join para que conserve los datos completos de user_usage
user_usage_device = pd.merge( left = user_usage,
                               right = user_device[['use_id', 'device']]
                               , how = 'left', on="use_id")
```

```python
# Mostrar las primeras cinco filas del DataFrame unido
user_usage_device.head(5)
```

**2.3.** Realizar el merge entre la tabla obtenida en el punto anterior y "android_devices", de manera que se preserven todos los datos de la tabla del punto anterior. 

Cuidar de no incluir en el merge variables innecesarias que no tengan relación con el problema. (15 puntos)

```python
# Merge entre las tablas user_usage_device y android_device
user_usage_android_device = pd.merge( left = user_usage_device
                               , right = android_device[['Retail Branding','Device']]
                               , how = 'left', left_on = 'device' , right_on = 'Device')
```

```python
# Mostrar las primeras cinco filas del DataFrame unido
user_usage_android_device.head(5)
```

```python
#Groupby RetailBranding,  outgoing_mins_per_month agrupar por marcas, y agregar por promedios: 'mean' 
user_usage_android_device.groupby(by="Retail Branding")["outgoing_mins_per_month"].mean()
```

```python
#Groupby RetailBranding, outgoing_sms_per_month agrupar por marcas, y agregar por promedios: 'mean'
user_usage_android_device.groupby(by="Retail Branding")["outgoing_sms_per_month"].mean()
```

```python
#Groupby RetailBranding, monthly_mb agrupar por marcas, y agregar por promedios: 'mean'
user_usage_android_device.groupby(by="Retail Branding")["monthly_mb"].mean()
```
